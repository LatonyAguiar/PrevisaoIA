# Projeto Ciência de Dados - Previsão de Vendas

- O objetivo é onseguir prever as vendas que vamos ter em determinado período com base nos gastos em anúncios nas 3 grandes redes que a empresa investe: TV, Jornal e Rádio
- TV, Jornal e Rádio estão em milhares de reais
- Vendas estão em milhões

#### Importar a Base de dados

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tabela = pd.read_csv("advertising.csv")
display(tabela)
```

```python
#Análise Exploratória
- Tentar visualizar como as informações de cada item estão distribuídas
- Ver a correlação entre cada um dos itens
```

```python
sns.heatmap(tabela.corr(), cmap='Wistia', annot=True)
plt.show()
```

#### Preparação dos dados para treinar o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
from sklearn.model_selection import train_test_split

# x = tabela[['TV', 'Radio', 'Jornal']] outra forma
x = tabela.drop('Vendas', axis=1)
y = tabela['Vendas']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

#### Problema de regressão - Escolher os modelos para usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

import numpy as np

#Criar e Treinar a IA
modelo_regressaolinear = LinearRegression()
modelo_regressaolinear.fit(x_train, y_train)

modelo_arvoredecisao = RandomForestRegressor()
modelo_arvoredecisao.fit(x_train, y_train)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
from sklearn import metrics

#Testar IA
previsao_regressaolinear = modelo_regressaolinear.predict(x_test)
previsao_arvoredacisao = modelo_arvoredecisao.predict(x_test)

r2_linear = metrics.r2_score(y_test, previsao_regressaolinear)
print(r2_linear)
r2_arvore = metrics.r2_score(y_test, previsao_arvoredacisao)
print(r2_arvore)
#rmse_lin = np.sqrt(metrics.mean_squared_error(y_test, test_pred_rf))
```

#### Visualização Gráfica das Previsões

```python
# a arvore de decisao teve maior porcentagem de previsão, então vou usar ela.

tabela_auxiliar = pd.DataFrame()
tabela_auxiliar['y_test'] = y_test
tabela_auxiliar['Previsao regressao linear'] = previsao_regressaolinear
tabela_auxiliar['Previsão arvoreDescisao'] = previsao_arvoredacisao

plt.figure(figsize=(15, 6))
sns.lineplot(data=tabela_auxiliar)
plt.show()
```

#### Como fazer uma nova previsão

```python
#Importar tabela com as novas informações apra prever

nova_tabela = pd.read_csv('novos.csv')
display(nova_tabela)
#usar modelo de arvore de decisao e fazer um .predict com ele
previsao = modelo_arvoredecisao.predict(nova_tabela)
print(previsao)
```
